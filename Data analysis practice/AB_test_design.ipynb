## Дизайн A/B теста: стоит ли внедрять новое размещение товаров

### Бизнес проблема и текущая ситуация: Сеть супермаркетов имеет три филиала в трех городах. Филиал - среднего размера супермаркет с товарами первой необходимости(базовыми),стандартными сувенирами и товарами премиум класса. Сейчас товары размещены так,что при входе клиент сначала видет сувениры и базовые товары,затем идут стенды с базовыми товарами и перпендикулярно им стоит стенд с премиум товарами. Вследствии изучения камер видеонаблюдения и статистики был сделан вывод о том, что доходя до стенда с премиум товарами,покупатель уже собирает нужную ему корзину,вследствие этого не рассматривая товары класса премиум.


### Нулевая гипотеза: Если внедрить новое размещение товаров,то выручка не изменится
### Альтернативная гипотеза: Если внедрить новое размещение товаров,то выручка изменится 

### Прибыль от решения проблемы: Добавив премиум товары и сувениры в стенды с обычными и переместив часть премиум товаров к стендам у кассы,конверсия в покупку этих товаров увеличится и вырчка вырастет от 5%

### Описание эксперимента: А/В тестирование будет проводиться во всех трех филиалах,первый месяц фиксируется выручка со старым размещением,следующие два месяца перемещаем премиум твоары и сувениры и фиксируем среднюю выручку,интерпретируем результаты.

### Метрики для анализа эксперимента: выручка с покупки(gross income)
