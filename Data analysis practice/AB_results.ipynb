```python
import pandas as pd
from scipy.stats import shapiro
from scipy.stats import levene
from scipy.stats import ttest_ind
import numpy as np
```

```python
df = pd.read_csv('supermarket_sales - Sheet1.csv.xls')
```

```python
df
```

```python
def dataframe_info(df):
    print(df.head(5))
    print(df.info()) 
    print(df.describe()) 
    print(df.isnull().sum())
    print(df.duplicated().sum())
```

```python
df['Date'].to_list()
```

```python
df['Date'] = df['Date'].astype('datetime64[ns]')
```

```python
df['Month'] = df['Date'].dt.month
```

```python
df_control = df[df['Month'] == 1]
df_test = pd.concat([df[df['Month'] == 2],df[df['Month'] == 3]])
```

```python
df_control
```

```python
contrin = df_control['gross income']
```

```python
testin = df_test['gross income']
```

```python
shapiro(df['gross income'])
```

### Принимаем нулевую гипотезу: среднее в выборках распределенно нормально

```python
levene(contrin,testin)
# Дисперсии равны
```

```python
ttest_ind(contrin, testin, equal_var=True)
```

```python
np.array(contrin).mean()
```

```python
np.array(testin).mean()
```

```python
print(np.median(np.array(contrin)))
print(np.median(np.array(testin)))
```

## После изучения метрики прибыли ,было выявлено что,между результатами тестовой и контрольной группы нет статистичсекой разницы,то есть мы принимаем нулевую гипотезу,и разворачивать эксперимент на всех клиентов не имеет смысла. Статистичсекой разницы нет,так как статистический критерий (парный Т-критерий) выше порога значимости в 0,05,значит принимается нулевая гипотеза,так же в выборках была проверена медиана, в медиане значимой разницы также нет.
